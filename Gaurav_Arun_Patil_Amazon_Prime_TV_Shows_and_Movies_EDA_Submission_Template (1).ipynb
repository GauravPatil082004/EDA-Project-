## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
# Remove duplicate rows from df_titles
initial_rows_titles = df_titles.shape[0]
df_titles.drop_duplicates(inplace=True)
duplicates_removed_titles = initial_rows_titles - df_titles.shape[0]
print(f"Removed {duplicates_removed_titles} duplicate rows from df_titles.")

# Remove duplicate rows from df_credits
initial_rows_credits = df_credits.shape[0]
df_credits.drop_duplicates(inplace=True)
duplicates_removed_credits = initial_rows_credits - df_credits.shape[0]
print(f"Removed {duplicates_removed_credits} duplicate rows from df_credits.")
````

Now, let's address the missing values. I will first handle the malformed entries (NaN in 'id', 'title', 'type') in `df_titles` by dropping those rows, as 'id' and 'title' are crucial identifiers. For other columns, I will fill missing values with appropriate placeholders.

````python
# Drop rows from df_titles where 'id' or 'title' are NaN (malformed entries)
initial_rows_titles = df_titles.shape[0]
df_titles.dropna(subset=['id', 'title'], inplace=True)
rows_dropped_malformed = initial_rows_titles - df_titles.shape[0]
print(f"Removed {rows_dropped_malformed} rows from df_titles due to missing 'id' or 'title'.")

# Fill missing 'description' with 'No description available'
df_titles['description'].fillna('No description available', inplace=True)

# Fill missing 'age_certification' with 'Unknown'
df_titles['age_certification'].fillna('Unknown', inplace=True)

# Fill missing 'seasons' with 0 for movies (type == 'MOVIE')
df_titles['seasons'] = df_titles.apply(lambda row: 0 if row['type'] == 'MOVIE' and pd.isna(row['seasons']) else row['seasons'], axis=1)

# For remaining NaN in 'seasons' (which should only be for SHOWs without season info), fill with median or mean if appropriate, or 0. For now, fill with 0.
df_titles['seasons'].fillna(0, inplace=True)

# Fill missing numerical columns with their median values
# imdb_score, imdb_votes, tmdb_popularity, tmdb_score
for col in ['imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']:
    if df_titles[col].isnull().any():
        median_val = df_titles[col].median()
        df_titles[col].fillna(median_val, inplace=True)
        print(f"Filled missing values in '{col}' with median: {median_val}")

# Fill missing 'imdb_id' with 'Unknown'
df_titles['imdb_id'].fillna('Unknown', inplace=True)


# For df_credits, fill missing 'character' with 'Unknown'
df_credits['character'].fillna('Unknown', inplace=True)
print("Filled missing values in 'character' column of df_credits with 'Unknown'.")


print('\nMissing values after wrangling in df_titles:')
display(df_titles.isnull().sum())
print('\nMissing values after wrangling in df_credits:')
display(df_credits.isnull().sum())
````

### What all manipulations have you done and insights you found?

**1. Duplicate Value Handling:**
- Removed 3 duplicate rows from `df_titles`.
- Removed 56 duplicate rows from `df_credits`.
*Insight*: Addressing duplicates ensures that each entry is unique and prevents skewed analyses caused by redundant data.

**2. Missing Value Handling in `df_titles`:**
- **'id' and 'title'**: Checked for and confirmed no missing values in these critical columns, meaning no rows were dropped for malformed entries.
- **'description'**: Filled 119 missing values with 'No description available'.
*Insight*: Providing a default string for missing descriptions maintains data completeness and allows for easier text analysis if needed.
- **'age_certification'**: Filled 6487 missing values with 'Unknown'.
*Insight*: This helps in categorizing content where certification is not available, preventing loss of data and allowing for analysis across all content.
- **'seasons'**: Filled missing values with 0. Specifically, for 'MOVIE' types where 'seasons' was NaN, it was set to 0. Any remaining NaNs in 'seasons' (which would typically be for 'SHOW' types with no season information) were also filled with 0.
*Insight*: This manipulation correctly distinguishes movies from shows and provides a numerical basis for 'seasons' for all entries, facilitating numerical analysis.
- **Numerical Columns (`imdb_score`, `imdb_votes`, `tmdb_popularity`, `tmdb_score`)**: Filled missing values with their respective median values.
    - `imdb_score`: Filled with median 6.1.
    - `imdb_votes`: Filled with median 464.0.
    - `tmdb_popularity`: Filled with median 2.537.
    - `tmdb_score`: Filled with median 6.0.
*Insight*: Using median imputation for numerical columns is robust to outliers and preserves the distribution of the data better than mean imputation, allowing for more accurate statistical analysis.
- **'imdb_id'**: Filled 667 missing values with 'Unknown'.
*Insight*: This ensures every title has an IMDb identifier, even if it's a placeholder, which can be useful for external lookups or analysis.

**3. Missing Value Handling in `df_credits`:**
